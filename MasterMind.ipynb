```python
#Importing Python libraries
import random
import itertools 
```

```python
#Counter for the number of tries
nb_tries = 0
 
#List of possible characters to be used in the code
char = ['Lion','Hippo','Bear','Monkey','Elephant','Tiger','Empty']

#Generating the list of still-possible answers--initially, all combinations are possible
posbl_answrs = list(itertools.permutations(char,3))

#Setting the code chosen by the codemaker
#Human-set
answr = ('Lion','Hippo','Bear')
#Code-set
#Generate a random integer from 1 to number of combinations--subtract 1 for Python indexing reasons
#rand = random.randint(0,len(posbl_answrs)-1)
#Take a random element from the list of combinations
#answr = posbl_answrs[rand]

#This variable will contain the code tried at each round
#Initialised to something stupid to enter the while loop
trial = ['Stupid','Stupid','Stupid']

#Repeat the loop while the guess does not match the code set by the codemaker
while (trial != answr):
    
    #Counter increment for the number of tries
    nb_tries += 1

    #Generate a random integer from 1 to number of still-possible answers
    #S ubtract 1 for Python indexing reasons
    idx = random.randint(0,len(posbl_answrs)-1)
    #Take a random element from the list of still-possible answers
    trial = posbl_answrs[idx]
    #Display the trial counter and the combination tried
    print('Try n°:',nb_tries,trial)

    #Count the number of obtained red and white dogs
    
    #For each of the three positions, compare the trial character with the answer character
    #If they are identical, add a red dog
    nbred = 0
    for i in range(3):
        if (trial[i] == answr[i]):
            nbred += 1
    #Display the number of red dogs obtained
    print('No of red dogs:',nbred)

    #For each of the three positions, compare the trial character with all three answer characters
    #If an identical character is found at a different position, add a white dog    
    nbwhite = 0
    for i in range(3):
        for j in range(3):
            if (trial[i] == answr[j] and i != j):
                nbwhite += 1
    #Display the number of white dogs obtained
    print('No of white dogs:',nbwhite)

    #Preparing to list all the still-possible answers after getting information from the dogs
    #We start with an empty list
    auxil = []

    #Go through the list of remaining possible answers allowed after the previous round
    #For each, we compare how many red and white dogs would have been obtained in the present round
    #had that possible answer been the actual code set by the codemaker
    #We retain the possible answers for which the number of red and white dogs
    #match the actually obtained number of red and white dogs
    
    for k in range(len(posbl_answrs)):
        #First count the number of hypothetical red dogs
        nbsubred = 0
        for i in range(3):
            if (trial[i] == posbl_answrs[k][i]):
                nbsubred += 1

        #Then count the number of hypothetical white dogs
        nbsubwhite = 0
        for i in range(3):
            for j in range(3):
                if (trial[i] == posbl_answrs[k][j] and i != j):
                    nbsubwhite += 1

        #If the number of red and white dogs matches, then the answer is still allowed
        #Here we thus add it to the list of the still-possible answers after this new round
        if (nbsubred == nbred and nbsubwhite == nbwhite):
            auxil.append(posbl_answrs[k])

    posbl_answrs = auxil
    
    #Display the number of still-possible answers after this round
    print('Possible answers:',len(posbl_answrs))
    
print("Congratulations! The code was solved!")
```
